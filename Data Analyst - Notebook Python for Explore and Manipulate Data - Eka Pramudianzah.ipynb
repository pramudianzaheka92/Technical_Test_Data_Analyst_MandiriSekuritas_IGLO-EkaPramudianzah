## 3. Dataset File "transactions_data,csv"

```python
data = pd.read_csv('/content/drive/MyDrive/Notes/File/Technical Test Mandiri Sekuritas/Dataset User Behavior/transactions_data.csv')

data.head()
```

```python
data["date"] = pd.to_datetime(data["date"])

data.dtypes
```

```python
data["year"] = data["date"].dt.year

data.head()
```

```python
# Menampilkan jumlah data yang hilang yang tersimpan di dalam file dataset "transactions_data.csv"
data.isnull().sum()
```

```python
# Menampilkan jumlah data duplikat yang tersimpan di dalam file dataset "transactions_data.csv"
data.duplicated().sum()
```

```python
import numpy as np

data_2019 = data[(data["year"] >= 2019) & (data["year"] <= 2019)]

data_2019.head()
```

```python
# Menghilangkan/menghapus kolom "errors" dan "zip" dari daftar kolom yang tersimpan di dalam file dataset "transactions_data.csv"
data_2019 = data_2019.drop(["errors", "zip"], axis = 1)

data_2019.head()
```

```python
# Menghapus hanya baris data yang memiliki/menyimpan nilai NaN (nilai missing value) yang masih tersimpan di dalam kolom "merchant_state"
data_2019 = data_2019.dropna(subset=['merchant_state'])

data_2019.isnull().sum()
```

```python
data_2019.to_csv("/content/drive/MyDrive/Notes/File/Technical Test Mandiri Sekuritas/Dataset User Behavior/transactions_data_2019.csv", index = False)
```
